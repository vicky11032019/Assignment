```
import numpy as np
import pandas as pd
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io

df2 = pd.read_excel(io.BytesIO(uploaded['mkcl age score data.xlsx']))
```

```
df2.head()
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,9))
plt.scatter(df2['Age'],df2['TotalScore'])
plt.show()
```

```
from sklearn.cluster import KMeans
```

```
km=KMeans(n_clusters=5)

km
```

```
y_pred=km.fit_predict(df2[['Age','TotalScore']])

y_pred
```

```
df2['cluster']=y_pred

df2.head()
```

```
import seaborn as sns

plt.figure(figsize=(12,9))
sns.scatterplot(x="Age", y="TotalScore",hue="cluster",data=df2)
plt.show()
```

```
knum=range(1,20)

sse=[]

for i in knum:
  km=KMeans(n_clusters=i)
  km.fit_transform(df2[['Age','TotalScore']])
  sse.append(km.inertia_)


```

```
plt.xlabel("K")
plt.ylabel("SSE")
plt.plot(sse)
plt.show()
```

elbow method to find which k value is good.
